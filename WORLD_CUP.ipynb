## **Model Building**

***CLASSIFICATION***

###***1.   Training the model***




Now, we’ll convert all “non-null object” columns to dummy variables using the following process:





```python
import pandas as pd

WINNER=pd.get_dummies(df.WINNER)
EDITION=pd.get_dummies(df.EDITION)
LOCATION=pd.get_dummies(df.LOCATION)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder

# Assuming 'df' is your DataFrame

# Create a OneHotEncoder object
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False for dense output

# Select categorical columns for encoding
categorical_cols = ['EDITION', 'LOCATION']  # Add other categorical columns if needed

# Fit and transform the encoder on the categorical columns
encoded_data = encoder.fit_transform(df[categorical_cols])

# Create a DataFrame from the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenate the encoded DataFrame with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)

# Drop the original categorical columns
df = df.drop(categorical_cols, axis=1)


# Now proceed with model training
y = df['WINNER']
X = df.drop('WINNER', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

model = LogisticRegression()
model.fit(X_train, y_train)
```

###***2.  Applying the model to make a prediction***



```python
y_model_train_pred = model.predict(X_train)
y_model_test_pred = model.predict(X_test)
```
